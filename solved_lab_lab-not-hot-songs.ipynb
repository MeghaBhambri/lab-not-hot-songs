```python
import pandas as pd
```

```python
#read csv file(not_top_100)
not_top_100 = pd.read_csv("genres_v2.csv")
not_top_100 = not_top_100.sample(10000)
display(not_top_100.head())
```

```python
#read csv file
top_100 = pd.read_csv("top_100.csv")
display(top_100.head())
```

```python
# remove the songs from "non_top_100" that are also in "top_100"
not_top_100 = not_top_100[~not_top_100['song_name'].isin(top_100['Songs'])]
```

```python
display(not_top_100.head())
```

```python
not_top_100.isna().sum()
```

```python
not_top_100 = not_top_100.dropna(subset=['song_name'])
```
